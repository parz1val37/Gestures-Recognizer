```python
import cv2 as cv
```

```python
# read image
img = cv.imread("spider.png")
```

```python
# resize image
img = cv.resize(img, (840, 620))
```

```python
img.shape
```

```python
cv.imshow("Spider", img)
# always write these two lines while viewing image or kernel fails
cv.waitKey(0)
cv.destroyAllWindows()
```

# Image(RGB) to Grayscale

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv.imshow("gray_scaled", gray_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# OpenCV is cv2
import cv2 as cv
```

```python
def WebCamfeed():
  cap = cv.VideoCapture(0) # Capture the webcam feed
  try:
    while True:
      res, frame = cap.read()
      if not res: break
      #flip the frame horizontally(1)
      frame = cv.flip(frame, 1)
      # show the frame
      cv.imshow('Feed', frame)

      if cv.waitKey(1) & 0xFF == ord('q'):
        break

  finally:
    cap.release()
    cv.destroyAllWindows()
```

```python
WebCamfeed()
```
